### 柱状图

```python
def cal_lis(name):
    temp = data_1[data_1['NAME'] == name]
    avl_income_country = [float(round(each /1000,2)) for each in temp['农村家庭可支配（元）'].tolist()]
    avl_income_city = [float(round(each /1000,2)) for each in temp['城市家庭可支配（元）'].tolist()]
    incomes_city = [float(round(each /1000,2)) for each in temp['城镇总收入（元）'].tolist()]
    incomes_country = [float(round(each /1000,2)) for each in temp['农村总收入（元）'].tolist()]
    return avl_income_country,avl_income_city,incomes_country,incomes_city
```

```python
def plot_line(timeData,avl_income,incomes,title,s1,s2):
    bar1 = (
        Bar()
        .add_xaxis(timeData)
        .add_yaxis(series_name=s1,y_axis=incomes)
        .extend_axis(
            yaxis=opts.AxisOpts(
                axislabel_opts=opts.LabelOpts(formatter="{value}"), interval=5
            )
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False),
                        itemstyle_opts=opts.ItemStyleOpts(opacity=0.8))
        .set_global_opts(
            tooltip_opts=opts.TooltipOpts(is_show=False),
            xaxis_opts=opts.AxisOpts(
                axislabel_opts={"interval":"0"}
              ),  
            yaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(formatter="{value}")),
        )
    )

    line1 = Line().add_xaxis(timeData).add_yaxis(
                            series_name=s2,
                            y_axis=avl_income,
                            symbol_size=8,
                            is_hover_animation=False,
                            label_opts=opts.LabelOpts(is_show=False),
                            linestyle_opts=opts.LineStyleOpts(width=1.5),
                            is_smooth=True,
                            yaxis_index=1
                        ).set_global_opts(
            title_opts=opts.TitleOpts(
                title=title, subtitle="2000年-2019年", pos_left="center"
            ),
            tooltip_opts=opts.TooltipOpts(trigger="axis"),
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True, link=[{"xAxisIndex": "all"}]
            ),
            datazoom_opts=[
                opts.DataZoomOpts(
                    is_show=True,
                    is_realtime=True,
                    start_value=30,
                    end_value=70,
                    xaxis_index=[0, 1],
                )
            ],
            xaxis_opts=opts.AxisOpts(
                type_="category",
                boundary_gap=False,
                axisline_opts=opts.AxisLineOpts(is_on_zero=True),
            ),
            yaxis_opts=opts.AxisOpts(name="元"),
            legend_opts=opts.LegendOpts(pos_left="left"),
            toolbox_opts=opts.ToolboxOpts(
                is_show=True,
                feature={
                    "dataZoom": {"yAxisIndex": "none"},
                    "restore": {},
                    "saveAsImage": {},
                },
            ),
        )
    bar1.overlap(line1)
    return bar1
```

```python
avl_income_country,avl_income_city,incomes_country,incomes_city = cal_lis(names[0])
chart1 = plot_line(years,avl_income_country,incomes_country,names[0]+'农村收入变化一览表','农村总收入','农村家庭可支配收入')
chart2 = plot_line(years,avl_income_city,incomes_city,names[0]+'城市收入变化一览表','城市总收入','城市家庭可支配收入')
tab1 = Tab()
tab1.add(chart1, names[0]+'农村收入变化一览表')
tab1.add(chart2, names[0]+'城市收入变化一览表')
tab1.render_notebook()
```

### 折线图

```python
def cal_lis1(name):
    temp = data_1[data_1['NAME'] == name]
    rate = [float(round(each,2)) for each in temp['城镇化率'].tolist()]
    population = [float(round(each,2)) for each in temp['城镇总人口（万人）'].tolist()]
    max_num = temp['城镇总人口（万人）'].max()
    return rate,population,max_num
```

```python
def plot_mul_line(s1,s2,title,timeData,rate1,rate2,rate3,population1,population2,population3,max_num):
    l1 = (
        Line()
        .add_xaxis(xaxis_data=timeData)
        .add_yaxis(
            series_name=s1[0],
            y_axis=population1,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .add_yaxis(
            series_name=s1[1],
            y_axis=population2,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .add_yaxis(
            series_name=s1[2],
            y_axis=population3,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=title, subtitle="2000年-2019年", pos_left="center"
            ),
            tooltip_opts=opts.TooltipOpts(trigger="axis"),
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True, link=[{"xAxisIndex": "all"}]
            ),
            datazoom_opts=[
                opts.DataZoomOpts(
                    is_show=True,
                    is_realtime=True,
                    start_value=30,
                    end_value=70,
                    xaxis_index=[0, 1],
                )
            ],
            xaxis_opts=opts.AxisOpts(
                type_="category",
                boundary_gap=False,
                axisline_opts=opts.AxisLineOpts(is_on_zero=True),
            ),
            yaxis_opts=opts.AxisOpts(min_=1000,max_=max_num, name="人"),
            legend_opts=opts.LegendOpts(pos_left="5%",pos_top='5%'),
            toolbox_opts=opts.ToolboxOpts(
                is_show=True,
                feature={
                    "dataZoom": {"yAxisIndex": "none"},
                    "restore": {},
                    "saveAsImage": {},
                },
            ),
        )
    )

    l2 = (
        Line()
        .add_xaxis(xaxis_data=timeData)
        .add_yaxis(
            series_name=s2[0],
            y_axis=rate1,
            xaxis_index=1,
            yaxis_index=1,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .add_yaxis(
            series_name=s2[1],
            y_axis=rate2,
            xaxis_index=1,
            yaxis_index=1,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .add_yaxis(
            series_name=s2[2],
            y_axis=rate3,
            xaxis_index=1,
            yaxis_index=1,
            symbol_size=8,
            is_hover_animation=False,
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=1.5),
            is_smooth=True,
        )
        .set_global_opts(
            axispointer_opts=opts.AxisPointerOpts(
                is_show=True, link=[{"xAxisIndex": "all"}]
            ),
            tooltip_opts=opts.TooltipOpts(trigger="axis"),
            xaxis_opts=opts.AxisOpts(
                grid_index=1,
                type_="category",
                boundary_gap=False,
                axisline_opts=opts.AxisLineOpts(is_on_zero=True),
                position="top",
            ),
            datazoom_opts=[
                opts.DataZoomOpts(
                    is_realtime=True,
                    type_="inside",
                    start_value=30,
                    end_value=70,
                    xaxis_index=[0, 1],
                )
            ],
            yaxis_opts=opts.AxisOpts(max_=1,is_inverse=True, name="%"),
            legend_opts=opts.LegendOpts(pos_left="10%",pos_top='90%'),
        )
    )

    grid = (
        Grid(init_opts=opts.InitOpts(width="1024px", height="768px"))
        .add(chart=l1, grid_opts=opts.GridOpts(pos_left=50, pos_right=50, height="35%"))
        .add(
            chart=l2,
            grid_opts=opts.GridOpts(pos_left=50, pos_right=50, pos_top="55%", height="35%"),
        )
    )
    return grid
```

```python
rate1,population1,max_num1 = cal_lis1(names[0])
rate2,population2,max_num2 = cal_lis1(names[1])
rate3,population3,max_num3 = cal_lis1(names[2])
max_num = max([max_num1,max_num2,max_num3])
s1, s2 = [], []
for each in names[:3]:
    s1.append(each + '城镇人口数')
    s2.append(each + '城镇化率')
line_data = plot_mul_line(s1,s2,'城镇人口数与城镇化率的关系',years,rate1,rate2,rate3,population1,population2,population3,max_num)
line_data.render_notebook()
```
